```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/huynhthanh98/ML/master/lab-06/bt1.csv')
```

```python
X = dataset.iloc[:,[3,4]].values
```

## 1. Dùng thư viện sch.dendrogram để vẽ ra dendrogram chọn ra số cụm thích hợp và dùng hierachical kmeans để gom cụm lại.
## 2. Vẽ hình biểu diễn các cụm

<ul>
    <li> <h3> Khoảng cách 'ward' </h3> </li>  
</ul>

```python
fig, ax = plt.subplots(figsize = (10, 8))
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'), ax = ax)
ax.axhline(y = 150, linestyle = 'dashed', color = 'k')
_ = ax.set(title = 'Dendogram', xlabel = 'Customers', ylabel = 'Euclid distance', xticks = [])
```

<h3> Từ biểu đồ trên, ta chọn k = 5 </h3>

```python
clustering = AgglomerativeClustering(linkage = 'ward', n_clusters = 5).fit(X)
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
labels = clustering.labels_
for id_label in np.unique(labels):
    points_cluster = X[labels == id_label]
    ax.scatter(points_cluster[:, 0], points_cluster[:, 1])
```
